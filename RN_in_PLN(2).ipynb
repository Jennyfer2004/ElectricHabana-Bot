```python
datos_solo_entidad = []

def generar_variantes(entidad):
    palabras = entidad.split()
    variantes = []

    # Generar todas las combinaciones de mayúsculas/minúsculas para la primera letra de cada palabra
    num_palabras = len(palabras)
    for i in range(2 ** num_palabras):  # Iterar sobre todas las combinaciones posibles
        nueva_palabra = []

        # Convertir el índice `i` en una lista de booleanos que indican si la letra debe ser minúscula
        cambios = [(i >> j) & 1 == 1 for j in range(num_palabras)]
        for j, palabra in enumerate(palabras):
            if cambios[j] and not palabra[0].islower():  # Cambiar la primera letra a minúscula
                if palabra[0].lower() + palabra[1:] in nueva_palabra:
                  continue
                nueva_palabra.append(palabra[0].lower() + palabra[1:])
            else:  # Mantener la primera letra como está
                nueva_palabra.append(palabra)

        if " ".join(nueva_palabra) in variantes:
          continue
        variantes.append(" ".join(nueva_palabra))
    return variantes

# Procesar los datos
i=0
for items in datos_solo_entidades:
    comentario=items[0]
    entidad=items[1]
    entidad_original = entidad[0]  # Entidad sin modificar

        # Generar variantes para la entidad
    variantes_entidad = generar_variantes(entidad_original)

        # Para cada variante, reemplazar la entidad en el comentario y añadir a la lista
    for variante in variantes_entidad:
            # Reemplazar la entidad original en el comentario por la variante
            comentario_modificado = comentario.replace(entidad_original, variante)
            datos_solo_entidad.append([comentario_modificado,variante])
            print(datos_solo_entidad[-1])
    i+=1
    print(i)
```

```python
import torch
import numpy as np
import torchvision
from transformers import AutoModel, AutoTokenizer, TrainingArguments, Trainer
from torch import nn
from datasets import Dataset, DatasetDict
import os
os.environ["WANDB_DISABLED"] = "true"
# 1. Cargar el tokenizador y los datos
tokenizer = AutoTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased")

def process_dataset(tokens, labels, tokenizer):
    encodings = tokenizer(
        tokens,
        is_split_into_words=True,
        padding=True,
        truncation=True,
        max_length=128,
        return_tensors="pt",
    )
    encoded_labels = []
    for i, label in enumerate(labels):
        word_ids = encodings.word_ids(batch_index=i)
        encoded_label = []
        previous_word_idx = None
        for word_idx in word_ids:
            if word_idx is None:
                encoded_label.append(-100)  # Ignorar tokens especiales
            elif word_idx != previous_word_idx:
                encoded_label.append(label_map[label[word_idx]])
            else:
                encoded_label.append(-100)  # Ignorar subpalabras
            previous_word_idx = word_idx
        encoded_labels.append(encoded_label)
    encodings["labels"] = encoded_labels
    return encodings


# Define un mapa para las etiquetas
label_map = {"O": 0, "B-LOC": 1, "I-LOC": 2}
examples_r=[]
labels_r=[]
for i,j in train_data_r:
  examples_r.append(i)
  labels_r.append(j)
val_labels_r=[]
val_data_=[]
for i,j in val_data_:
  val_data_.append(i)
  val_labels_r.append(j)
train_encodings = process_dataset(examples_r, labels_r, tokenizer)
val_encodings = process_dataset(val_data_, val_labels_r, tokenizer)

# Convertir a dataset de Hugging Face
from torch.utils.data import Dataset

class NERDataset(Dataset):
    def init(self, encodings):
        self.encodings = encodings

    def len(self):
        return len(self.encodings["input_ids"])

    def getitem(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

train_dataset = NERDataset(train_encodings)
val_dataset = NERDataset(val_encodings)


# 2. Definir el modelo BERT con capas adicionales
class CustomBERTNER(nn.Module):
    def init(self, model_name, num_labels):
        super(CustomBERTNER, self).init()
        self.bert = AutoModel.from_pretrained(model_name, ignore_mismatched_sizes=True)
        self.dropout = nn.Dropout(0.3)
        self.linear = nn.Linear(self.bert.config.hidden_size, 256)  # Capa extra
        self.relu = nn.ReLU()
        self.classifier = nn.Linear(256, num_labels)  # Capa final para clasificación

    # Inicializar la función de pérdida con ponderaciones de clase
        self.loss_fn = nn.CrossEntropyLoss(weight=manual_weights_tensor, ignore_index=-100)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = outputs.last_hidden_state
        x = self.dropout(sequence_output)
        x = self.linear(x)
        x = self.relu(x)
        logits = self.classifier(x)

        loss = None
        if labels is not None:

            # Calcular la pérdida utilizando la función de pérdida ponderada
            loss = self.loss_fn(logits.view(-1, logits.shape[-1]), labels.view(-1))

        return {"loss": loss, "logits": logits}
```

```python
def predict_entities(text, model, tokenizer, id_to_label):
    # Tokenizar el texto de entrada
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]

    # Obtener las predicciones
    model.eval()
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=-1)

    # Convertir IDs de etiquetas a nombres
    tokens = tokenizer.convert_ids_to_tokens(input_ids.squeeze())
    predicted_labels = [id_to_label[label_id.item()] for label_id in predictions[0]]

    # Mostrar tokens con sus etiquetas
    entities = []
    for token, label in zip(tokens, predicted_labels):
        if label.startswith("B-LOC"):
            entities.append(token)

    return entities

# Ejemplo de predicción
text = "Fui a Madrid y después viajé a París."
entities = predict_entities(text, model, tokenizer, id_to_label)
print("Entidades encontradas:", entities)
```

```python
import torch
import numpy as np
from transformers import AutoModelForTokenClassification, AutoTokenizer
from datasets import Dataset, DatasetDict

# 1. Definir el mapeo de etiquetas
label_map = {"O": 0, "B-LOC": 1, "I-LOC": 2}
label_map_inv = {v: k for k, v in label_map.items()}

# 2. Cargar el tokenizador y el modelo entrenado

# 3. Lista de textos nuevos para predecir
new_texts = [
    "Voy a viajar a Santiago de Chile.",
    "Barcelona es una ciudad hermosa.",
    "Madrid es la capital de España.",
    "Pero es que Versalles aqui en la Lisa ya es mucho ....hay ancianos que ni siquiera pueden cobrar en ese Banco y hoy es 13 ....la verdad es bien complicado"

]

# 4. Tokenizar los textos nuevos
new_encodings = tokenizer(
    new_texts,
    is_split_into_words=False,  # Los textos no están tokenizados previamente
    padding=True,
    truncation=True,
    max_length=128,
    return_tensors="pt",  # Devolver tensores de PyTorch
)

# 5. Realizar predicciones
with torch.no_grad():
    outputs = modelo(input_ids=new_encodings["input_ids"], attention_mask=new_encodings["attention_mask"])
    logits = outputs["logits"]  # Obtener los logits del modelo

# 6. Aplicar softmax y obtener predicciones
probs = torch.softmax(logits, dim=-1).numpy()
preds = [
    [np.argmax(p) if max(p) > 0.4 else 0 for p in seq]  # Umbral personalizado
    for seq in probs
]

# 7. Decodificar las predicciones
for i, pred_seq in enumerate(preds):
    tokens = tokenizer.convert_ids_to_tokens(new_encodings["input_ids"][i])
    previous_word_idx = None
    current_entity = []  # Para almacenar palabras de una entidad
    entities = []  # Para almacenar todas las entidades encontradas

    words = new_texts[i].split()  # Dividir el texto en palabras originales

    for word_idx, pred_label in zip(word_ids, pred_seq):
        if word_idx is not None and word_idx != previous_word_idx:
            # Verificar que word_idx esté dentro del rango de las palabras originales
            if word_idx < len(words):
                token = words[word_idx]  # Obtener la palabra original
                label = label_map_inv[pred_label]

                if label == "B-LOC":
                    # Comenzar una nueva entidad
                    if current_entity:
                        entities.append(" ".join(current_entity))
                        current_entity = []
                    current_entity.append(token)
                elif label == "I-LOC":
                    # Continuar la entidad actual
                    current_entity.append(token)
                else:
                    # Finalizar la entidad si existe
                    if current_entity:
                        entities.append(" ".join(current_entity))
                        current_entity = []

        previous_word_idx = word_idx

    # Asegurarse de agregar la última entidad si existe
    if current_entity:
        entities.append(" ".join(current_entity))

    # Imprimir resultados
    print("Texto:", new_texts[i])
    print("Entidades detectadas:", entities)
    print()
```
